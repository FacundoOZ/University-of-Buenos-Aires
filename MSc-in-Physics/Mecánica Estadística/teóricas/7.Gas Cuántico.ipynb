# Gas cuántico
En este notebook aprendemos a crear funciones en Python, y las usamos para graficar la densidad de ocupación de cada nivel de energía en un gas de partículas clásicas, de fermiones, y de bosones, a diferentes temperaturas. Recordar que cada celda se debe ejecutar en orden, apretando SHIFT+ENTER:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})

# Definimos funciones para la ocupación media de cada nivel de energía
def dens_quantum(ener,kT,a):
	return 1/(np.exp(ener/kT)+a)

def dens_classic(ener,kT):
	return np.exp(-ener/kT)
```

Definimos el potencial químico (en unidades arbitrarias), y construimos un vector con diferentes valores de la energía, para poder hacer gráficos.

```python
mu   = 10                      # Potencial químico (en unidades de una energía epsilon_0)
ener = np.arange(1000)/50 - mu # Vector con la energía menos el potencial químico
```

Vemos primero la ocupación para un gas de partículas clásicas (Boltzmann):

```python
plt.figure(figsize=(8,5))
plt.plot(ener,dens_classic(ener,20),label=r'$kT = 20 \epsilon_0$')
plt.plot(ener,dens_classic(ener,10),label=r'$kT = 10 \epsilon_0$')
plt.plot(ener,dens_classic(ener,5),label=r'$kT = 5 \epsilon_0$')
plt.legend(loc="upper right")
plt.xlabel(r'$(\epsilon - \mu)/\epsilon_0$')
plt.ylabel(r'$n_\epsilon$')
plt.show()
```

Para Fermiones:

```python
a = 1 # Fermiones

plt.figure(figsize=(8,5))
plt.plot(ener,dens_quantum(ener,10,a),label=r'$kT = 10 \epsilon_0$')
plt.plot(ener,dens_quantum(ener,5,a),label=r'$kT = 5 \epsilon_0$')
plt.plot(ener,dens_quantum(ener,1,a),label=r'$kT = 1 \epsilon_0$')
plt.plot(ener,dens_quantum(ener,.5,a),label=r'$kT = 0.5 \epsilon_0$')
plt.plot(ener,dens_quantum(ener,.05,a),label=r'$kT = 0.05 \epsilon_0$')
plt.legend(loc="upper right")
plt.xlabel(r'$(\epsilon - \mu)/\epsilon_0$')
plt.ylabel(r'$n_\epsilon$')
plt.show()
```

Para Bosones:

```python
a    = -1           # Bosones
ene2 = ener + 1e-10 # Desplazo los valores de energía en un número pequeño, para evitar dividir por cero

plt.figure(figsize=(8,5))
plt.plot(ene2,dens_quantum(ene2,100,a),label=r'$kT = 100 \epsilon_0$')
plt.plot(ene2,dens_quantum(ene2,20,a),label=r'$kT = 20 \epsilon_0$')
plt.plot(ene2,dens_quantum(ene2,10,a),label=r'$kT = 10 \epsilon_0$')
plt.plot(ene2,dens_quantum(ene2,5,a),label=r'$kT = 5 \epsilon_0$')
plt.plot(ene2,dens_quantum(ene2,1,a),label=r'$kT = 1 \epsilon_0$')
plt.plot(ene2,dens_quantum(ene2,.05,a),label=r'$kT = 0.05 \epsilon_0$')
plt.legend(loc="upper right")
plt.ylim(0,100)                                                         # Limitamos el rango del eje y porque la densidad crece mucho a baja temperatura
plt.xlabel(r'$(\epsilon - \mu)/\epsilon_0$')
plt.ylabel(r'$n_\epsilon$')
plt.show()
```

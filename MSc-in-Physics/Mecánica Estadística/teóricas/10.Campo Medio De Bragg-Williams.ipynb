# Campo medio en la aproximación de Bragg-Williams
En este notebook calculamos el valor de los términos que aparecen en la sumatoria de la aproximación de Bragg-Williams, y graficamos el resultado. Recuerden ejecutar cada celda aprentando SHIFT+ENTER:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

# Fijamos un valor para N, y creamos un vector con valores para N+ y L
N  = 100              # Número de spines
Np = np.arange(0,N,1) # N+ toma valores enteros entre 0 y N
L  = 2*Np/N-1         # L (el "orden macroscópico" o de "largo rango")

# Fijamos valores para beta, q, J, y mu.B
beta = 1   # 1/kT
q    = 4   # Número de primeros vecinos
J    = .1  # Energía de interacción entre vecinos
muB  = 1   # Producto del momento dipolar magnético mu y el campo externo B

# Calculamos un vector con todos los términos de la sumatoria
term = factorial(N)/(factorial(Np)*factorial(N-Np)) * np.exp(beta*N*L*(q*L*J/2 + muB))
```

```python
plt.figure(figsize=(7, 5))                                                # Graficamos los términos de la suma en función de L
plt.plot(L, term)
plt.xlabel(r'$L$', fontsize=14)
plt.ylabel(r'$N!/[N_+ (N-N_+)!] e^{\beta NL(qLJ/2+\mu B)}$', fontsize=14)
plt.show()
```

Veamos el error que cometemos al calcular el logaritmo de la suma de todos los términos, contra la aproximación en la que calculamos el logaritmo del término más grande:

```python
np.log(np.sum(term)) # Calculamos el logaritmo de la suma de todos los términos
```

```python
np.log(np.max(term)) # Calculamos el logaritmo del término más grande
```

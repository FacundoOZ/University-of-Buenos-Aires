# Formación de gotas
Este notebook muestra cómo graficar una función usando Python. En particular, para el problema de condensación graficamos la presión de saturación en función del radio de la gota, y el radio de la gota en función de la presión. Para ejecutarlo, apreten SHIFT+ENTER en cada celda (en orden, comenzando por la primera). En Google Colaboratory también pueden ejecutar las celdas aprentando el ícono de "play" arriba a la izquierda de cada celda (aparece al poner el dedo o el puntero del mouse sobre los corchetes "[ ]" que dan inicio a cada celda con código). Comenzamos importando NumPy (https://numpy.org/devdocs/user/quickstart.html) y Matplotlib (https://matplotlib.org/3.3.1/tutorials/index.html), y definiendo algunos parámetros para el problema de condensación.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos parámetros para el problema de condensación:
gamma = 72e-3         # Tensión superficial del agua [N/m]
m     = 0.018/6.02e23 # Masa de una molécula de agua, 18 g por mol [kg]
rho   = 997           # Densidad del agua [kg/m^3]
T     = 247           # Temperatura en la parte superior de una nube [K]
k     = 1.38e-23      # Constante de Boltzmann [m^2 kg/(s^2 K)]
```

Generamos un vector con diferentes valores posibles para el radio de la gota, y graficamos la presión dividida por la presión de saturación P/Ps en función del radio r. Para eso usamos la función plot de Matplotlib. Si cambian valores en esta celda, o en la celda previa, recuerden apretar SHIFT+ENTER para volver a ejecutar cada una de las celdas:

```python
r   = (np.arange(10000)+1)*1e-8     # Creamos un vector con números crecientes entre 1 y 10000 y lo multiplicamos por 1e-8
pps = np.exp(2*gamma*m/(rho*k*T*r)) # Calculamos P/Ps

plt.figure()     # Graficamos los resultados
plt.semilogx(r,pps)
plt.xlabel(r'$r$ [m]')
plt.ylabel(r'$P/P_s$')
plt.show()
```

Ahora graficamos la función inversa. Generamos un vector con posibles valores de la presión en unidades de la presión de saturación (P/Ps), comenzando desde 1 (pues P/Ps >= 1). Luego calculamos los valores de r para los diferentes valores de la presión, y graficamos el resultado:

```python
pps = 1 + (np.arange(10000)+1)*1e-5 # Generamos un vector con P/Ps con valores >= 1
r = 2*gamma*m/(rho*k*T*np.log(pps)) # Calculamos r

plt.figure()     # Graficamos los resultados
plt.semilogy(pps,r)
plt.xlabel(r'$P/P_s$')
plt.ylabel(r'$r$ [m]')
plt.show()
```

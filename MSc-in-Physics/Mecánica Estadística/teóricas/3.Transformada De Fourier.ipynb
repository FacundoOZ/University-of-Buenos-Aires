# Transformada de Fourier
Calculamos la transformada Fourier de una función cuadrada centrada en el cero usando SymPy (https://docs.sympy.org/latest/tutorial/index.html). Luego usamos NumPy (https://numpy.org/devdocs/user/quickstart.html) y Matplotlib (https://matplotlib.org/tutorials/) para graficar los resultados en forma numérica. Recuerden ejecutar cada celda aprentando SHIFT+ENTER:

```python
s,kappa = symbols("s kappa",real=True)         # Definimos dos variables simbólicas reales s y kappa
simplify(integrate(exp(I*kappa*s)/2,(s,-1,1))) # Calculamos la transformada Fourier de una función cuadrada
```

Graficamos numéricamente la función cuadrada y su transformada Fourier usando MumPy y Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos un vector para las posiciones en x y k
x = np.arange(-5,5,.001)   # x toma valores entre  -5 y 5 , con un paso de 0.001
k = np.arange(-20,20,.001) # k toma valores entre -20 y 20, con un paso de 0.001

plt.figure()                                               # Graficamos la función cuadrada f(x) (construida como producto de funciones de Heaviside)
plt.plot(x,.5*(1-np.heaviside(x-1,0))*np.heaviside(x+1,0))
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.show()
plt.figure()                                               # Graficamos la transformada Fourier Q(k)
plt.plot(k,np.sin(k)/k)
plt.xlabel(r'$k$')
plt.ylabel(r'$Q(k)$')
plt.show()
```

Veamos ahora que pasa al tomar potencias cada vez más grandes de Q(k). El valor de N se puede cambiar para aumentar la potencia:

```python
N = 40                           # Prueben aumentar este número de a poco y volver a ejecutar esta celda (SHIFT+ENTER)

plt.figure()                     # Graficamos (Q(k))^N en escala lineal
plt.plot(k,(np.sin(k)/k)**N)
plt.xlabel(r'$k$')
plt.ylabel(r'$Q^N(k)$')
plt.show()
plt.figure()                     # Graficamos (Q(k))^N en escala semilog
plt.semilogy(k,(np.sin(k)/k)**N)
plt.xlabel(r'$k$')
plt.ylabel(r'$Q^N(k)$')
plt.show()
```
